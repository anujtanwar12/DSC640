# Plots Using **Python**

## Load Data

```python
# Load lib raries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data
crimeDF = pd.read_csv("./ex4-2/crimerates-by-state-2005.csv")
birthDF = pd.read_csv("./ex4-2/birth-rate.csv")

# Remove outliers i.e. Washington DC in our dataset
crimeDF = crimeDF[crimeDF.state != 'District of Columbia']
# And remove United States
crimeDF = crimeDF[crimeDF.state != 'United States']
```

<div style="page-break-after: always;"></div>

## Scatterplot

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 6))

# Create a Scatterplot
plt.plot('murder', 'burglary', data=crimeDF, linestyle='none', markersize=10, 
         markerfacecolor='b', marker='X', markeredgecolor="none")

# Add chart title and labels
plt.title("Python Scatter Plot - Murders vs Burglaries in the United States (excluding DC)", fontsize = 15, loc = 'left')
plt.xlabel("Murders", fontsize = 12)
plt.ylabel("Burglaries", fontsize = 12)

plt.show()
```

<div style="page-break-after: always;"></div>

## Bubble Chart

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(20, 10))

# Make a bubble plot
sc = plt.scatter('murder', 'burglary', s=crimeDF.population*.0001, data=crimeDF, marker='o',
                 alpha=0.7, color='b', label='Population')

# Add chart title and labels
plt.title("Python Bubble Chart - Murders vs Burglaries in the United States (excluding DC)", fontsize = 24, loc = 'left')
plt.xlabel("Murders", fontsize = 20)
plt.ylabel("Burglaries", fontsize = 20)

# Set legend for population sizes                   
handles, labels = sc.legend_elements(prop="sizes", alpha=0.4, num=3)     
labels = ["10 MM", "20 MM", "30 MM"]     
legend = ax.legend(handles, labels, loc="lower right", title="Population", markerscale=0.3,
                   fontsize=11, title_fontsize=14) 

plt.show()
```

<div style="page-break-after: always;"></div>

## Density Plot

```python
# Initialize the matplotlib figure
f, ax = plt.subplots(figsize=(10, 6))

# Make a density plot
sns.kdeplot(birthDF['1980'], color='b', fill='b')

# Add chart title and labels
plt.title("Python Density Plot - Global Distribution of Birth Rates in 1980", fontsize = 15, loc = 'left')
plt.xlabel("Live Births per 1,000 Population", fontsize = 12)
plt.ylabel("Density", fontsize = 12)
```
