```python
# Import libraries
import csv
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime as dt
```

```python
# Read world population data
population = pd.read_excel('ex2-2/world-population.xlsm',0)

print(population.head())
```

### World Population data
#### Line and Step graph

```python
X = population['Year']
Y = population['Population']
plt.plot(X, Y)
plt.title('Line - Python (Population Vs Year)')
plt.show()
```

```python
plt.step(X, Y)
plt.title('Stepped - Python (Population Vs Year)')
plt.show()
```
